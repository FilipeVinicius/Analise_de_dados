#2. Cálculo de 'm' (Número de Votos no Quantil 80%):
Atribua à variável 'm' o cálculo do número de votos dos filmes no quantil de 80%. Utilize a coluna 'vote_count' para isso.


```python
df["vote_count"]
```

```python
m = df["vote_count"].quantile(0.8)
```

```python
m
```

# 3. Filtragem por Duração dos Filmes:
Considere apenas filmes com duração entre 45
minutos e 300 minutos. Utilize a coluna 'runtime' para aplicar esse filtro.

```python
filmes45_300 = df[(df["runtime"] >= 45) & (df["runtime"] <= 300)]
filmes45_300
```

```python
filmes45_300.describe()
```

# 4. Filtragem por Número de Votos:
Considere apenas filmes que receberam mais de 'm'
votos. Use a coluna 'vote_count' novamente para essa filtragem.


```python
filmes45_300_m = filmes45_300[filmes45_300["vote_count"] >= m]
filmes45_300_m
```

# 5. Contagem de Filmes após Filtragem:
Verifique quantos filmes passaram pelos filtros.

```python
filmes45_300_m.shape
```

# 6. Cálculo da Média de 'vote_average':
Atribua à variável 'C' a média da coluna
'vote_average'. A coluna 'vote_average' representa a média das avaliações dadas pelos
espectadores ou usuários para um determinado filme.


```python
C = df["vote_average"].mean()
C
```

# 7. Função para Classificação Ponderada:
Crie uma função chamada
'classificacao_ponderada' para calcular a classificação ponderada de cada filme. Utilize a
função fornecida na questão.


```python
def classificacao_ponderada(x, m=m, C=C):
  v = x['vote_count']
  R = x["vote_average"]
  return (v/(v+m) * R) + (m/(m+v)*C)
```

# 8. Cálculo da Pontuação e Adição de Coluna 'score':
 Calcule a pontuação de cada filme
e adicione uma nova coluna chamada 'score' ao DataFrame usando a função
'classificacao_ponderada' definida na questão 7.

```python
filmes45_300_m["score"] = filmes45_300_m.apply(classificacao_ponderada, axis=1)
filmes45_300_m["score"]
```



# 9. Ordenação por Pontuação:

Ordene os filmes em ordem decrescente com base em suas pontuações


```python
ordenar_decrescente = filmes45_300_m.sort_values(by="score", ascending=False)
```

```python
ordenar_decrescente["score"]
```

```python
ordenar_decrescente
```
